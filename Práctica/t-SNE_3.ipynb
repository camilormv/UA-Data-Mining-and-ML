```python
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler

# Cargamos el conjunto de datos de dígitos
digits = datasets.load_digits()

# Normalizamos los datos
scaler = StandardScaler()
data_scaled = scaler.fit_transform(digits.data)

#Aplicamos t_SNE
tsne = TSNE(n_components=2, random_state=0)
data_tsne = tsne.fit_transform(data_scaled)

#Visualizamos los resultados
plt.figure(figsize=(10, 8))
colors = ["#476A2A", "#7851B8", "#BD3430", "#4A2D4E" , "#875525",
          "#A83683", "#4E655E", "#853541", "#3A3120", "#535D8E"]
for i in range(len(data_tsne)):
    # Coloca los digitos en el scatter plot en la posición correspondiente 
    plt.scatter(data_tsne[i, 0], data_tsne[i, 1], c=colors[digits.target[i]])
plt.show()
```
